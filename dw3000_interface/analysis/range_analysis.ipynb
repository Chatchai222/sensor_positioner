```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]

plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
sample_data = pd.read_csv('0_65m_chatchaiHouse_idealCondition.csv')


ranges_error = sample_data.range - 0.65
plt.title('DW3000 0.65m Chatchai house, ideal condition')
plt.xlabel("range error:")
plt.ylabel("count")
plt.hist(ranges_error, bins=20, edgecolor="black")
plt.show()
print("ranges:")
print("standard deviation: ", ranges_error.std())
print("average: ", ranges_error.mean())

ranging_attempt_counts = sample_data.ranging_attempt_count
plt.title("DW3000 0.65m Chatchai house, ideal condition")
plt.xlabel("ranging_attempt")
plt.ylabel("count")
plt.hist(ranging_attempt_counts, bins=20, edgecolor="black")
plt.show()
print("ranging_attempt: ")
print("data point amount: ", len(ranging_attempt_counts))
print("standard deviation: ", ranging_attempt_counts.std())
print("average: ", ranging_attempt_counts.mean())
```

```python
sample_data = pd.read_csv("2_60m_chatchaiHouse_idealCondition_0_30mOffGround_90degRotatePortrait.csv")

ranges_error = sample_data.range - 2.60
plt.xlabel("range error(meter)")
plt.ylabel("count")
plt.hist(ranges_error, bins=100, edgecolor="black")
plt.show()
print("ranges error:")
print("standard deviation: ", ranges_error.std())
print("average: ", ranges_error.mean())

ranging_attempt_counts = sample_data.ranging_attempt_count
plt.xlabel("ranging_attempt")
plt.ylabel("count")
plt.hist(ranging_attempt_counts, bins=100, edgecolor="black")
plt.show()
print("ranging_attempt: ")
print("data point amount: ", len(ranging_attempt_counts))
print("standard deviation: ", ranging_attempt_counts.std())
print("average: ", ranging_attempt_counts.mean())
print("max: ", ranging_attempt_counts.max())
```
