```python
import pandas as pd
import math
from matplotlib import pyplot as plt

def plot_ranging_data(csv_file, distance):
    csv_data = pd.read_csv(csv_file)
    
    ranges_error = csv_data.range - distance
    plt.title(csv_file)
    plt.xlabel("range error(meter)")
    plt.ylabel("count")
    plt.hist(ranges_error, bins=50, edgecolor="black")
    plt.show()
    
    square_ranges_error = ranges_error ** 2
    mean_square_ranges_error = square_ranges_error.mean()
    root_mean_square_ranges_error = math.sqrt(mean_square_ranges_error)

    absolute_ranges_error = ranges_error.abs()
    
    print("range error: ")
    print("total data point: ", len(ranges_error))
    print("root mean square: ", root_mean_square_ranges_error)
    print("standard deviation: ", ranges_error.std())
    print("max: ", ranges_error.max())
    print("min: ", ranges_error.min())
    print("absolute 99th percentile: ", absolute_ranges_error.quantile(0.99, interpolation="nearest"))
    

    ranging_attempt_counts = csv_data.ranging_attempt_count
    plt.title(csv_file)
    plt.xlabel("ranging attempt")
    plt.ylabel("count")
    plt.hist(ranging_attempt_counts, bins=50, edgecolor="black")
    plt.show()

    print("ranging attempt: ")
    print("total data point: ", len(ranging_attempt_counts))
    print("standard deviation: ", ranging_attempt_counts.std())
    print("average: ", ranging_attempt_counts.mean())
    print("99th percentile: ", ranging_attempt_counts.quantile(0.99, interpolation="nearest"))
    print("max: ", ranging_attempt_counts.max())
    
    

plot_ranging_data("0_65m_chatchaiHouse_directLineOfSight_0_00mOffGround_landscapeMirror.csv", 0.65)
plot_ranging_data("1_20m_ECC706_directLineOfSight_0_00mOffGround_portraitMirror.csv", 1.20)
plot_ranging_data("1_30m_chatchaiHouse_directLineOfSight_0_00mOffGround_landscapeMirror.csv", 1.30)
plot_ranging_data("2_40m_ECC706_aluminiumPlateInBetween_0_10mOffGround_portraitMirror.csv", 2.40)
plot_ranging_data("2_40m_ECC706_directLineOfSight_0_00mOffGround_portraitMirror.csv", 2.40)
plot_ranging_data("2_40m_ECC706_directLineOfSight_0_10mOffGround_portraitMirror.csv", 2.40)
plot_ranging_data("2_60m_chatchaiHouse_aluminumFoilCoverSensor_0_30mOffGround_portraitMirror.csv", 2.60)
plot_ranging_data("2_60m_chatchaiHouse_directLineOfSight_0_30mOffGround_landscapeMirror.csv", 2.60)
plot_ranging_data("2_60m_chatchaiHouse_directLineOfSight_0_30mOffGround_portraitMirror.csv", 2.60)
plot_ranging_data("2_60m_chatchaiHouse_directLineOfSight_0_30mOffGround_ZAxisPierceMirror.csv", 2.60)
plot_ranging_data("3_60m_ECC706_directLineOfSight_1_50mOffGround_ZAxisPierceMirror.csv", 3.60)
plot_ranging_data("6_00m_chatchaiHouse_directLineOfSight_0_30mOffGround_landscapeMirror.csv", 6.00)
plot_ranging_data("6_00m_chatchaiHouse_directLineOfSight_0_30mOffGround_portraitMirror.csv", 6.00)
plot_ranging_data("9_00m_chatchaiHouse_directLineOfSight_0_30mOffGround_landscapeMirror.csv", 9.00)
plot_ranging_data("9_00m_chatchaiHouse_directLineOfSight_0_30mOffGround_portraitMirror.csv", 9.00)

```
